2.  **`gTTS()`**: This takes our text and creates the actual speech sound.
3.  **`tts.save()`**: This saves the generated speech as an audio file (an MP3).
4.  **`Audio()`**: This plays the saved audio file right here in Colab so you can hear it.

```python
from gtts import gTTS
from IPython.display import Audio
```

```python
text_to_speak = "Hello! I am your own voice assistant"
tts = gTTS(text=text_to_speak, lang='en')
tts.save("voice_assistant.mp3")
Audio("voice_assistant.mp3", autoplay=True)
```
